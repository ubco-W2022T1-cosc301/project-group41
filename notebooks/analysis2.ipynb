### Task 2: Set up an “Analysis Pipeline”

#### 1. Load Data

We have already done so in the beginning of the file.

#### 2. Clean Data

Since we have 3 years worth of consecutive time series data, we will first clean and drop the columns which we do not need.
We are going to base our research on the economy and trust in the government, and hence we'll keep those and their related columns in our set.

```python
clean_data1 = data1.drop(columns=['Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', "Generosity"])

clean_data2 = data2.drop(columns=['Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', "Generosity"])

clean_data3 = data3.drop(columns=['Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', "Generosity"])
```

Now we will rename the columns in each set to match the year in which they were recorded, and shorten their names for easy access.

```python
clean_data1.rename(columns={"Happiness Score": "2015 Score", "Economy (GDP per Capita)": "2015 GDP", "Trust (Government Corruption)": "2015 Trust", "Dystopia Residual": "2015 Dystopia Residual"}, inplace=True)

clean_data2.rename(columns={"Happiness Score": "2016 Score", "Economy (GDP per Capita)": "2016 GDP", "Trust (Government Corruption)": "2016 Trust", "Dystopia Residual": "2016 Dystopia Residual"}, inplace=True)

clean_data3.rename(columns={"Happiness Score": "2017 Score", "Economy (GDP per Capita)": "2017 GDP", "Trust (Government Corruption)": "2017 Trust", "Dystopia Residual": "2017 Dystopia Residual"}, inplace=True)
```

Then we will merge these datasets into one frame which gives us a time series of information for each country.

```python
timeseries = pd.merge(clean_data1, clean_data2, how="left", on="Country")
timeseries = pd.merge(timeseries, clean_data3, how="left", on="Country")

timeseries = timeseries.dropna() #Used to drop any rows which had null values

timeseries.head()
```

#### 3. Process Data

For this section we will first create a column for the average Dystopia Residual of the 3 years. Dystopia Residual is the hypothetical happiness score of the country with the worst values across its parameters, it is used in order to set a benchmark for comparision.

```python
timeseries["AvgDR"] = ((timeseries["2015 Dystopia Residual"] + timeseries["2016 Dystopia Residual"] + timeseries["2017 Dystopia Residual"]) / 3)

timeseries = timeseries.drop(columns=["2015 Dystopia Residual", "2016 Dystopia Residual", "2017 Dystopia Residual"])

timeseries.head()
```

### Task 4: Conduct your analysis to help answer your research question(s)

With our EDA we will now conduct a thorough analysis of our research question(s).

```python
relation = timeseries.corr(numeric_only=True)
sns.heatmap(relation, annot=True, cmap="magma", linecolor="white", linewidths=1)
```

According to our **correlation coefficient** of **0.78, 0.81, and 0.82** stated in the above heat map, we can conclude that a **stable economy (GDP per Capita)** does have a **positive impact** on the **happiness** of the country and its people. A wealthier country will have more resources and therefore can provide for their citizens better than a financially unstable country.

Meanwhile, we noticed that the correlation of **0.42, 0.44, and 0.45** do **not have a significant influence** on the **happiness score**, which is suprising since the peoples trust in the government should be one of the major factors which affects their happiness as the government is the legal body which is responsible for enforcing new laws and upholding previous ones.

Let us analyse how the average dystopia residual relates to the Trust in the government;

```python
sns.set_palette("magma")
s = sns.PairGrid(timeseries, y_vars=["AvgDR"], x_vars=["2015 Trust", "2016 Trust", "2017 Trust"])
s.map(sns.kdeplot)
```

From the above contour plot we can conclude that the **destopia residual** is in **higher concentration** in places of **lower trust** in the government. But this **does not mean** that they **influence one another** as the plot values for **DR are spread between the 1 to 3 range**. Furthermore, from the **heat map factors of 0.02, 0.015, and 0.014** we can confidently state that **trust and the dystopia residual are unrelated**.
