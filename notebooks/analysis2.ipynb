### Task 1: EDA

First, we'll get familiar with our dataset.

Since all 3 of our datasets have similar columns I am going to work on one dataframe for the sake of simplicity and do a joint analysis in the further tasks.

```python
print("2015 data:", data1.shape)
print("2016 data:", data2.shape)
print("2017 data:", data3.shape)

data1.describe()
```

```python
sns.set_theme(style="dark")
```

Let us find out how different columns in our dataset correlate with each other;

```python
relation = data1.corr(numeric_only=True)
relation
```

```python
sns.heatmap(relation, annot=True, cmap="afmhot", linecolor="white", linewidths=1)
```

In the above heatmap we notice an intresting finding where the **economy** of the country has a **high correlation** with **happiness score** compared to the other parameters.

```python
sns.jointplot(x="Economy (GDP per Capita)", y='Happiness Score', data=data1, kind="reg")
```

As we can see here, it is obvious that the **higher the economy** of the country the **greater is the happiness score**, it is linearly dependant.

```python
sns.jointplot(x="Economy (GDP per Capita)", y='Happiness Score', hue="Trust (Government Corruption)", data=data1)
```

According to this plot we notice that the **economy** and the peoples **trust in the government** are somewhat related to one another as a **greater trust** also implies a **better economy (a few exceptions)**.

### Task 2: Set up an “Analysis Pipeline”

Since we have 3 years worth of consecutive time series data we will first clean and drop the columns which we do not need.
We are going to base our research on the economy and trust in the government, and hence we'll keep those and their related columns in our set.

```python
data1 = data1.drop(columns=['Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', "Generosity"])

data2 = data2.drop(columns=['Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', "Generosity"])

data3 = data3.drop(columns=['Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', "Generosity"])

#timeseries = pd.DataFrame([data1["Country"], data1["Happiness Score"], data1["Economy (GDP per Capita)"], data1["Trust (Government Corruption)"]]).transpose()
#timeseries.head()
```

Now we will rename the columns in each set to match the year in which they were recorded.

```python
data1.rename(columns={"Happiness Score": "2015 Happiness Score", "Economy (GDP per Capita)": "2015 Economy (GDP per Capita)", "Trust (Government Corruption)": "2015 Trust (Government Corruption)", "Dystopia Residual": "2015 Dystopia Residual"})

data2.rename(columns={"Happiness Score": "2016 Happiness Score", "Economy (GDP per Capita)": "2016 Economy (GDP per Capita)", "Trust (Government Corruption)": "2016 Trust (Government Corruption)", "Dystopia Residual": "2016 Dystopia Residual"})

data3.rename(columns={"Happiness Score": "2017 Happiness Score", "Economy (GDP per Capita)": "2017 Economy (GDP per Capita)", "Trust (Government Corruption)": "2017 Trust (Government Corruption)", "Dystopia Residual": "2017 Dystopia Residual"})
```
